```python
# I HAVE CREATED THE COLOR DISTRIBUTION CHART AND THE DOMINANT COLOR IN THE PICTURE TOO.
# REFERENCE ATTACHED AT THE END OF THIS CODE SNIPPET

#Import dependencies
import cv2
import numpy as np

#function to create color distribution chart 
def create_bar(height, width, color):
    bar = np.zeros((height, width, 3), np.uint8)
    bar[:] = color
    #bgr to rgb color format
    red, green, blue = int(color[2]), int(color[1]), int(color[0])
    return bar, (red, green, blue)

#Read any image "make sure format is correct"
img = cv2.imread('desktop/bali_small.jpeg')
#Extract parameters from image
height, width, _ = np.shape(img)
# print(height, width)

#Extract all pixel points and color channels from image
data = np.reshape(img, (height * width, 3))
data = np.float32(data)

#Using Kmeans Clustering algorithm to make random clusters around the image "used 3 clusters here but you can adjust as many"
number_clusters = 3
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
flags = cv2.KMEANS_RANDOM_CENTERS
compactness, labels, centers = cv2.kmeans(data, number_clusters, None, criteria, 10, flags)
print(centers)



#Here I have written code snippet for detecting the most dominant color in image
#Theory is I have first extracted the cluster having color which is repeated mostly. 'dense cluster of all'
y=max(centers, key = lambda x: x[0])
# Now I have repeated the process 'done below for showing color distribution chart' for that dense cluster
barss = []
rgb_valuess = []

bar, rgbb = create_bar(200, 200, y)
barss.append(bar)
rgb_valuess.append(rgbb)

img_barrr = np.hstack(barss)
font = cv2.FONT_HERSHEY_SIMPLEX
for y in enumerate(rgb_valuess):
    image = cv2.putText(img_barrr, 'The dominant color ^', (10 , 180),
                        font, 0.5, (255, 0, 0), 1, cv2.LINE_AA)
    image = cv2.putText(img_barrr, f'RGB{y[1]}', (10 , 160),
                        font, 0.5, (255, 0, 0), 1, cv2.LINE_AA)




#For showing color distribution chart
font = cv2.FONT_HERSHEY_SIMPLEX
bars = []
rgb_values = []

#loop through the cluster points and fill the bar with respective values
for index, row in enumerate(centers):
    bar, rgb = create_bar(200, 200, row)
    bars.append(bar)
    rgb_values.append(rgb)
#hstack will stack all values horizontly, here we have done the color distribution chart.
img_bar = np.hstack(bars)

#To put rgb values to respective bars using cv2.putText
for index, row in enumerate(rgb_values):
    image = cv2.putText(img_bar, f'{index + 1}. RGB:{row}', (5 + 200 * index, 200 - 10),
                        font, 0.5, (255, 0, 0), 1, cv2.LINE_AA)
    print(f'{index + 1}. RGB{row}')

#To show final output
cv2.imshow('Image', img)
out= np.concatenate((img_bar,img_barrr), axis=1)
cv2.imshow('output', out)
#cv2.imshow('dominate_color', img_barrr)
#cv2.imshow('other dom', img_bar)

# cv2.imwrite('output/bar.jpg', img_bar)

cv2.waitKey(0)
cv2.destroyAllWindows()


#Reference : https://www.youtube.com/watch?v=90s4SomOSa0
```
